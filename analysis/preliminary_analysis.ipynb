# Results

## Variables in data

- **PassengerId** Unique identifier, no relationship to Survivability
- **Survived** Information on survival, *0=No, 1=Yes*
- **Pclass** Ticket class, *1=1st class, 2=2nd class, 3=3rd class*, **extremely** relevant
- **Name** Passenger name, on 1st approximation *ir*relevant, could be used for nationality determination which might be relevant
- **Sex** Passenger Sex, **extremely** relevant
- **Age** Passenger Age in years, relevant, null-value handling to be determined
- **SibSp** number of siblings / spouses aboard the Titanic, relevant
- **Parch** number of parents / children aboard, relevant, should most likely be transformed to two different variables based on age
- **Ticket** Ticket number, **most likely** irrelevant to be analyzed if model doesn't perform properly without it
- **Fare** Passenger fare, relevant
- **Cabin** Cabin number. ***Many empty values -> these are highly relevant***, **Letter value** might be relevant -> to be analysed, **multiple cabins** might be relevant -> to be analysed. 
- **Embarked** Port of Embarkation, seems to be relevant.
